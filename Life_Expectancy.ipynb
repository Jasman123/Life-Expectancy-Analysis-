```python
def show_data_EDA(df) :
  #read number of column
  sum_data = len(df.columns)

  #initiating figure table
  fig, ax = plt.subplots(sum_data, 2, figsize=(12, 5 * sum_data))

  #read all data through for loop
  for i in range(sum_data):
      col_name = df.columns[i]

    # Left plot - Histogram with KDE
      sns.histplot(data=df, x=col_name, kde=True, ax=ax[i, 0])
      ax[i, 0].set_title(f'Histogram of {col_name}')

    # Right plot - Boxplot
      sns.boxplot(data=df, x=col_name, ax=ax[i, 1])
      ax[i, 1].set_title(f'Boxplot of {col_name}')

  plt.tight_layout()
  plt.show()
```

```python
#Making summarize of missing data
df.isnull().sum()
```

```python
#Checking correlation of each columm to predicting value ( life expactancy)
#need to filter out the categorical data ( number only)
df_numeric = df.select_dtypes(include=np.number)
df_numeric.head()
```

```python
# Initiating corr_life data frame that contain correlation of each variable
corr_life = df_numeric.corr()

#Fillter out the major correlation to 'life expectancy in which has to be over 0,5 or below -0.5
corr_life = corr_life[(corr_life['Life expectancy '] >= 0.5) | (corr_life['Life expectancy '] <= -0.5)]
features =corr_life['Life expectancy '].index.tolist()
features
```

With variable configuration, it concludes that in numeric variable six column have major inffluence to life expectancy or prediction variable

```python
#Merge Catergorical Features
df_1 = df[features]
df_1 = pd.concat([df_1, df[['Country','Year','Status']]], axis =1)
df_1.head()
```

```python
df.info()
```

```python
#Summ of Nan or null Data in each columns
df_1.isnull().sum()
```

```python
#drop NaN value in minor missing value
df_1.dropna(subset=['Adult Mortality','Life expectancy ',' BMI '], axis=0, inplace=True)

#Fill NaN value with mean or average
avg_income = df_1['Income composition of resources'].astype('float').mean(axis=0)
df_1['Income composition of resources'].replace(np.nan,avg_income, inplace=True)

avg_school = df_1['Schooling'].astype('float').mean(axis=0)
df_1['Schooling'].replace(np.nan, avg_school, inplace=True)


df_1.dropna(axis=0, inplace=True)

df_1.isnull().sum()
```

```python
#analysis graph of each data statically
show_data_EDA(df_1)
```

```python
#Income composition of resources possible multicolinearity with 'Schooling'
df_3 = df_2.drop('Income composition of resources', axis = 1)
df_3.head()
```

```python
columns_to_encode = ['Country', 'Year','Status']
df_dummies = pd.get_dummies(df_3, columns=columns_to_encode, drop_first=True)
df_dummies.head()
```

```python
df_dummies = df_dummies.astype(float)
df_dummies.head()
```

```python
df_dummies.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df_dummies.drop('Life expectancy ',axis=1)
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

Y = df_dummies['Life expectancy ']

x_train, x_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=3)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("RÂ² Score:", r2_score(y_test, y_pred))
```

```python
for col, coef in zip(X.columns, model.coef_):
    print(f"{col}: {coef:.3f}")
```

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('predict value')
plt.ylabel('residuals')
plt.title('Residual Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, edgecolor='k')
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--')
plt.grid(True)
plt.show()
```

```python
coff_data = pd.DataFrame(data=model.coef_, index=X.columns, columns=['Coefficient'])
coff_data = coff_data.reset_index()
coff_data = coff_data.rename(columns={'index': 'Feature'})
coff_data.head()
```

```python
df_country = coff_data[coff_data['Feature'].str.contains('Country', case=False)]
df_country = df_country.rename(columns={'Feature': 'Country'})
df_country['Country'] = df_country['Country'].str.replace('Country_','')
df_country.reset_index(drop=True, inplace=True)
df_country.sort_values(by='Coefficient', ascending=False).reset_index(drop=True).head(10)
```

```python
df_year = coff_data[coff_data['Feature'].str.contains('year', case=False)]
df_year = df_year.rename(columns={'Feature' : 'Year'})
df_year['Year'] = df_year['Year'].str.replace('Year_','')
df_year.reset_index(drop=True, inplace=True)
df_year.head()

plt.figure(figsize=(10, 6))
plt.plot(df_year['Year'], df_year['Coefficient'], marker='o', linestyle='--')
plt.title('Change of Coefficient each years', fontsize=16)
plt.xlabel('year', fontsize=12)
plt.ylabel('Coefficient', fontsize=12)
plt.grid(True)
plt.show()
```

```python
df_status = coff_data[~(coff_data['Feature'].str.contains('year', case=False) | coff_data['Feature'].str.contains('country', case=False))]
df_status.reset_index(drop=True, inplace=True)
df_status.head()
```

```python
coefs = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

# Plot top 10 positive and negative coefficients
plt.figure(figsize=(10, 6))
top_features = pd.concat([coefs.head(10), coefs.tail(10)])
top_features.set_index('Feature')['Coefficient'].plot(kind='barh', color='skyblue')
plt.title("Linear Regression - Top 10 Positive and Negative Coefficients")
plt.axvline(0, color='gray', linewidth=0.8)
plt.tight_layout()
plt.show()
```
